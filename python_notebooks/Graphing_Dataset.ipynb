```python
import pandas as pd
df = pd.read_csv('Kaggle_Books-Selling-Records-.csv')
df
```

# Making a barplot
This will just be a sample barplot, the goal is to product a script that will return this plot after inputting the dataset (.csv file)

```python
import matplotlib.pyplot as plt 
df1 = df.reset_index()["Qtr"].value_counts()
ax = df1.nlargest(10).sort_values(ascending=True).plot.barh(x='Quater', y='Number of Sales', title= 'Number of Sales per Quater')
ax.set(xlabel='Number of Sales', ylabel='Quaters')
plt.show()
```
